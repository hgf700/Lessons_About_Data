# Wizualizacja Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Wykres liniowy

```python
# Dane
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Tworzenie wykresu
plt.plot(x, y, marker='^', linestyle='-', color='b', label="Liniowy")
plt.xlabel("Oś X")
plt.ylabel("Oś Y")
plt.title("Wykres Liniowy")
plt.legend()
plt.show()
```

## Wykres słupkowy

```python
# Dane
kategorie = ['A', 'B', 'C', 'D']
wartosci = [5, 7, 3, 8]

# Tworzenie wykresu słupkowego
plt.bar(kategorie, wartosci, color='orange')
plt.xlabel("Kategorie")
plt.ylabel("Wartości")
plt.title("Wykres Słupkowy")
plt.show()
```

## Histogram

```python
# Dane
dane = np.random.normal(0, 1, 1000)

# Tworzenie histogramu
plt.hist(dane, bins=30, color='purple', alpha=0.7)
plt.xlabel("Wartości")
plt.ylabel("Częstotliwość")
plt.title("Histogram")
plt.show()
```

## Wykres kołowy

```python
# Dane
kategorie = ['A', 'B', 'C', 'D']
wartosci = [15, 30, 45, 10]

# Tworzenie wykresu kołowego
plt.pie(wartosci, labels=kategorie, autopct='%1.1f%%', startangle=90)
plt.title("Wykres Kołowy")
plt.show()
```

## Kolore, linie i style markerów

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

plt.plot(x, y, color='green', marker='x', linestyle='--', linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Dostosowany Wykres Liniowy")
plt.show()
```

## Dodawanie siatki

```python
plt.plot(x, y, color='blue', marker='o')
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Wykres z Siatką")
plt.show()
```

## Dodawanie adnotacji

```python
plt.plot(x, y, marker='o')
plt.annotate('Najwyższy punkt', xy=(5, 35), xytext=(3, 30),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.title("Wykres z Adnotacją")
plt.show()
```

## Wykres punktowy

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Wykres Punktowy")
plt.show()
```

## Wykres 3d

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='plasma')
plt.title("Wykres 3D")
plt.show()
```

# Biblioteka plotly

```python
# Dane
kategorie = ['A', 'B', 'C', 'D']
wartosci = [4, 7, 1, 8]

# Tworzenie wykresu słupkowego
fig = px.bar(x=kategorie, y=wartosci, title="Wykres Słupkowy", labels={'x':"Kategorie", 'y':"Wartości"})
fig.show()
```

```python
# Dane
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
colors = np.random.rand(50)

fig = px.scatter(x=x, y=y, size=sizes, color=colors, title="Wykres Punktowy")
fig.show()
```

```python
import plotly.graph_objects as go

# Dane
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Tworzenie wykresu 3D
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
fig.update_layout(title="Wykres 3D", scene=dict(
                    xaxis_title='X',
                    yaxis_title='Y',
                    zaxis_title='Z'))
fig.show()
```

## Wykresy z subplotami

```python
from plotly.subplots import make_subplots

# Dane
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 17, 10]
y2 = [5, 6, 2, 3, 8]

# Tworzenie subplotów
fig = make_subplots(rows=1, cols=2, subplot_titles=("Wykres 1", "Wykres 2"))
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines+markers', name="Liniowy 1"), row=1, col=1)
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines+markers', name="Liniowy 2"), row=1, col=2)
fig.update_layout(title="Subploty w Plotly")
fig.show()
```

# Wprowadzenie do biblioteki plotnine

```python
from plotnine import *
import pandas as pd

# Przykładowe dane
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [2, 4, 6, 8, 10]
})
```

```python
# Tworzenie podstawowego wykresu
(ggplot(data) + aes(x='x', y='y') + geom_point())
```

```python
# Wykres punktowy
(ggplot(data) + aes(x='x', y='y') + geom_point() +
 ggtitle("Wykres Punktowy") +
 xlab("Oś X") + ylab("Oś Y"))
```

```python
# Wykres liniowy
(ggplot(data) + aes(x='x', y='y') + geom_line(color='blue') +
 ggtitle("Wykres Liniowy") +
 xlab("Oś X") + ylab("Oś Y"))
```

```python
# Dane do wykresu słupkowego
data_bar = pd.DataFrame({
    'kategorie': ['A', 'B', 'C', 'D'],
    'wartości': [4, 7, 1, 8]
})

# Wykres słupkowy
(ggplot(data_bar) + aes(x='kategorie', y='wartości') +
 geom_bar(stat='identity', fill='skyblue') +
 ggtitle("Wykres Słupkowy") +
 xlab("Kategorie") + ylab("Wartości"))
```

```python
import numpy as np

# Dane
data_hist = pd.DataFrame({
    'wartości': np.random.normal(0, 1, 1000)
})

# Histogram
(ggplot(data_hist) + aes(x='wartości') +
 geom_histogram(bins=30, fill='purple', alpha=0.7) +
 ggtitle("Histogram") +
 xlab("Wartości") + ylab("Częstotliwość"))
```

```python
# Wykres punktowy z dostosowanymi kolorami
(ggplot(data) + aes(x='x', y='y') +
 geom_point(color='red', size=3) +
 ggtitle("Wykres Punktowy z Dostosowanymi Kolorami") +
 xlab("Oś X") + ylab("Oś Y"))
```

```python
# Wykres punktowy z linią trendu
(ggplot(data) + aes(x='x', y='y') +
 geom_point() + geom_smooth(method='lm') +
 ggtitle("Wykres z Linią Trendu") +
 xlab("Oś X") + ylab("Oś Y"))
```

```python
# Przykładowe dane
data_facet = pd.DataFrame({
    'x': np.tile([1, 2, 3, 4], 2),
    'y': [1, 2, 3, 4, 2, 3, 4, 5],
    'grupa': ['A']*4 + ['B']*4
})

# Wykres z facetowaniem
(ggplot(data_facet) + aes(x='x', y='y') +
 geom_point() + facet_wrap('~grupa') +
 ggtitle("Wykres z Facetowaniem") +
 xlab("Oś X") + ylab("Oś Y"))
```

```python
# Przykładowe dane
data_box = pd.DataFrame({
    'kategorie': np.random.choice(['A', 'B', 'C'], 100),
    'wartości': np.random.randn(100)
})

# Wykres pudłowy
(ggplot(data_box) + aes(x='kategorie', y='wartości') +
 geom_boxplot(fill='lightblue') +
 ggtitle("Wykres Pudłowy") +
 xlab("Kategorie") + ylab("Wartości"))
```

```python
# Dodanie motywu
(ggplot(data) + aes(x='x', y='y') + geom_point() +
 ggtitle("Wykres z Motywem") +
 theme_minimal())
```

# Wizualizacja Geoprzestrzenna Danych za Pomocą Plotly w Pythonie

```python
import plotly.express as px
import pandas as pd

# Przykładowe dane miast z ich współrzędnymi
data = pd.DataFrame({
    'city': ['New York', 'London', 'Tokyo', 'Sydney'],
    'latitude': [40.7128, 51.5074, 35.6895, -33.8688],
    'longitude': [-74.0060, -0.1278, 139.6917, 151.2093]
})

# Tworzenie mapy punktowej
fig = px.scatter_mapbox(data, lat="latitude", lon="longitude", hover_name="city",
                        zoom=1, height=500)

# Konfiguracja stylu mapy
#fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(mapbox_style="carto-positron")
fig.update_layout(title="Lokalizacja Wybranych Miast na Mapie")
fig.show()
```

```python
import plotly.express as px

# Dane krajów z ich kodami ISO oraz populacją
data = px.data.gapminder().query("year == 2007")

# Tworzenie mapy choropleth
fig = px.choropleth(data, locations="iso_alpha", color="pop",
                    hover_name="country", color_continuous_scale="Viridis",
                    labels={'pop': 'Populacja'})

# Ustawienia mapy
fig.update_layout(title="Populacja Krajów na Świecie w 2007 roku")
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

np.random.seed(0)
lats = np.random.normal(37.7749, 0.05, 500)
lons = np.random.normal(-122.4194, 0.05, 500)
z_weights = np.random.uniform(1, 10, 500)

fig = go.Figure(go.Densitymapbox(
    lat=lats,
    lon=lons,
    z=z_weights,
    radius=25,
    colorscale='Hot'
))

fig.update_layout(
    mapbox_style="open-street-map",
    mapbox_center_lon=-122.4194,
    mapbox_center_lat=37.7749,
    mapbox_zoom=10,
    title="Mapa Cieplna Przedstawiająca Gęstość Punktów"
)

fig.show()
```

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'State': ['CA', 'TX', 'FL', 'NY', 'PA', 'IL', 'OH', 'GA', 'NC', 'MI'],
    'Population': [39538223, 29145505, 21538187, 20201249, 13002700, 12812508, 11799448, 10711908, 10439388, 10077331]
})

# 2. Tworzenie mapy choropleth
fig = px.choropleth(df,
                    locations="State",           # Kolumna ze skrótami (np. 'CA')
                    locationmode="USA-states",   # Informujemy Plotly, że to stany USA
                    color="Population",          # Kolumna z wartościami
                    color_continuous_scale="Viridis",
                    labels={'Population': 'Populacja'})

# 3. Ustawienia mapy
fig.update_layout(title="Populacja wybranych Stanów USA", geo_scope='usa')
fig.show()
```
