```python
# wy´swietli´c ´srednia (maksymalna˛, minimalna˛) warto´s´c z jednej kolumny

df["value_max"] = df["value"].max()
df["value_min"] = df["value"].min()
df["value_mean"] = df["value"].mean()

df[["value_max","value_min","value_mean"]]
```

```python
# sortowanie wierszy ramki danych wedlug warto´sci okre´slonej kolumny
# (maleja˛co, rosna˛co)

df["value1"] = df["value"].sort_values(ascending=False).values
df["value2"] = df["value"].sort_values(ascending=True).values
df[["value1","value2"]]
```

```python
# wy´swietli´c wierszy dla 10 najwie˛kszych (najmniejszych) warto´sci okre´slonej
# kolumny

dflarg=df.nlargest(10,'value')
dfsmall=df.nsmallest(10,'value')

dflarg
dfsmall
```

```python
# • wy´swietli´c wierszy dla 10 najwie˛kszych warto´sci okre´slonej kolumny
# pod warunkiem okre´slonych warto´sci innej kolumny

top10 = df[(df["value"] == 1.0) & (df["answer_label"] == "Male")].nlargest(10, "value")
top10
```

```python
# grupowanie wierszy wedlug warto´sci kolumny kategoryzowanej, potem
# - u´srednienie warto´sci wszystkich kolumn w grupie - MultiIndex

df_group = df.groupby(['question', 'answer_label']).agg({'value': ['mean', 'median']})
df_group.xs('Are you alone right now?',level='question')
```

```python
# grupowanie wierszy wedlug warto´sci kolumny kategoryzowanej, potem 
# - u´srednienie warto´sci dla pewnych kolumn, liczba warto´sci i mediana 
# dla pozostalych kolumn w grupach

df_group2 = df.groupby(['question', 'answer_label']).agg({
    'value': ['mean', 'median'],
    'question_type':['unique']
})
df_group2
```
